```python
!pip install pandas


import pandas as pd

# Specify the path to your CSV file
csv_file_path = 'team_stats_2003_2023.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

df.head()
```

```python


print(df.info())
print(df.describe())
```

```python
df.columns = df.columns.str.strip()
selected_columns = ['team', 'pass_att', 'wins', 'win_loss_perc']
df = df[selected_columns]

print(df)
print(df[['team', 'pass_att', 'wins', 'win_loss_perc']].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.scatterplot(x='pass_att', y='wins', data=df)
plt.title('Scatter Plot of Pass Attempts vs Wins')
plt.xlabel('Pass Attempts')
plt.ylabel('Wins')
plt.show()
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.histplot(df['pass_att'], kde=True)
plt.title('Distribution of Pass Attempts')

plt.subplot(2, 2, 2)
sns.histplot(df['wins'], kde=True)
plt.title('Distribution of Wins')

plt.subplot(2, 2, 3)
sns.histplot(df['win_loss_perc'], kde=True)
plt.title('Distribution of Win-Loss Percentage')

plt.tight_layout()
plt.show()
```
